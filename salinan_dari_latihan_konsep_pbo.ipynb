# Pendekatan Prosedural

 Pengelompokan satu tugas tertentu yang bisa digunakan berkali-kali sebagai pendekatan dalam pemecahan suatu masalah.

```python
def hitung_luas ():
  alas = float(input('Masukkan alas: '))
  tinggi = float(input('Masukkan tinggi: '))
  print('Luas =', 0.5 * alas * tinggi)
```

```python
# kita bisa panggil berkali-kali
hitung_luas()
# panggil lagi
hitung_luas()
```

# Pendekatan Fungsional

Hampir sama dengan prosedural, hanya saja paradigma ini lebih berorientasi terhadap “input-output” dari pada mengubah data secara langsung seperti pada paradigma prosedural.

```python
def input_alas_dan_tinggi ():
  alas = float(input('Masukkan alas: '))
  tinggi = float((input('Masukkan tinggi: ')))
  return alas, tinggi

def hitung_luas (alas, tinggi):
  return 0.5 * alas * tinggi

# kalau fungsional, kita sendiri yang mengelola hasil kembaliannya
```

```python
# satu fungsi bisa dipanggil secara independen
print(hitung_luas(5, 10))

# contoh dengan inputan alas dan tinggi
alas, tinggi = input_alas_dan_tinggi()
print(hitung_luas(alas, tinggi))
```

# Pendekatan Berorientasi Obyek

Semua komponen program seolah-olah adalah sebuah objek. Sebuah objek selalu memiliki identitas dan juga perilaku atau kemampuan untuk melakukan tugas tertentu.

![objek-segitiga[1].png]()



```python
class Segitiga:
  def __init__(self, alas, tinggi):
    self.alas = alas
    self.tinggi = tinggi

  def get_luas(self):
    return 0.5 * self.alas * self.tinggi
```

```python
# Proses membuat obyek dari class (instansiasi)
s1 = Segitiga(5, 10)
s2 = Segitiga(10, 10)

print('luas segitiga1:', s1.get_luas())
print('luas segitiga2:', s2.get_luas())
```
